```python
import pandas as pd
import json

def read_jsonl(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield json.loads(line)

data = list(read_jsonl('sampleds4.jsonl'))
df = pd.json_normalize(data)

print(df)
```
